```python
# Load the country shapefile as a GeoDataFrame:
shapes = gpd.read_file('../ne_10m_admin_0_countries.zip')
shapes = shapes.rename(columns={'ADMIN': 'name'})
shapes = shapes[['SOV_A3', 'name', 'geometry']]
shapes
```

```python
df_country_region = pd.DataFrame(wb.economy.list())
df_country_region.head()
```

```python
regions = df_country_region[df_country_region['aggregate'] == True]
regions.head()
```

```python
df_country =  df_country_region[df_country_region['aggregate'] == False]
df_country.head()
```

```python
df_country_shape = df_country.merge(shapes, left_on='value', right_on='name', how='left')
df_country_shape[:25]
```

```python
for index, row in df_country_shape.iterrows():
    if pd.isna(row['name']):        
        # Ensure idx is a single row, not a DataFrame
        idx = shapes.loc[shapes['SOV_A3'] == row['id']]
        
        if not idx.empty:  # Check if a match is found
            df_country_shape.loc[index, ['name', 'geometry']] = idx.iloc[0][['name', 'geometry']]

df_country_shape
```
